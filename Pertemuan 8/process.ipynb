```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
import numpy as np
```

```python
#load dataset
file_path = 'produksiPadi.xlsx'
data = pd.read_excel(file_path, sheet_name='data training')
```

```python
data.head()
```

```python
#lagging data
data['Lagging Data -1'] = data['Produksi Padi(Ton)'].shift(1)
data['Lagging Data -2'] = data['Produksi Padi(Ton)'].shift(2)
data['Lagging Data -3'] = data['Produksi Padi(Ton)'].shift(3)
lagged_data = data.dropna()
```

```python
lagged_data.head()
```

```python
#split feature x dan y
X = lagged_data[['Lagging Data']]
y = lagged_data['Produksi Padi(Ton)']
```

```python
# split 70% training 30% testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, shuffle=False)
```

```python
# Regresi
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
#prediksi
y_pred = model.predict(X_test)
```

```python
model.score(X_test, y_test)
```

```python
#evaluasi mse, rmse, mape
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mape = mean_absolute_percentage_error(y_test, y_pred)
```

```python
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
print("Mean Absolute Percentage Error (MAPE):", mape)
```
